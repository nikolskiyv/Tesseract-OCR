```python
from docx import Document
import stanza
import zipfile
import xml.etree.ElementTree
import os
```

```python
WORD_NAMESPACE = '{http://schemas.openxmlformats.org/wordprocessingml/2006/main}'
PARA = WORD_NAMESPACE + 'p'
TEXT = WORD_NAMESPACE + 't'
TABLE = WORD_NAMESPACE + 'tbl'
ROW = WORD_NAMESPACE + 'tr'
CELL = WORD_NAMESPACE + 'tc'
```

```python
stanza.download('en')
stanza.download('ru')
```

```python
def stanza_nlp_ru(text):
    #f = open('ru.txt', 'a')
    nlp = stanza.Pipeline(lang='ru', processors='tokenize,ner')
    doc = nlp(text)
    for sent in doc.sentences: 
        for ent in sent.ents:  
            #if ent.type == 'ORG':
                #f.write(ent.text + '\n')
            print(ent.text + ' ' + ent.type)
    #f.close()            
```

```python
def stanza_nlp_en(text):
    #f = open('en.txt', 'a')
    nlp = stanza.Pipeline(lang='en', processors='tokenize,ner')
    doc = nlp(text)
    for sent in doc.sentences: 
        for ent in sent.ents:  
            #if ent.type == 'ORG':
            #f.write(ent.text + '\n')
            print(ent.text + ' ' + ent.type)
    #f.close()        
```

```python
k = 0
for root, dirs, files in os.walk("."):  
    for filename in files:
        if filename.endswith("docx") and k < 10:
            print(filename)
            k = k+1
            
            my_text = ''

            with zipfile.ZipFile(filename) as docx:
                tree = xml.etree.ElementTree.XML(docx.read('word/document.xml'))

            for table in tree.iter(TABLE):
                for row in table.iter(ROW):
                    for cell in row.iter(CELL):
                        my_text += ''.join(node.text for node in cell.iter(TEXT))
                        my_text += '\n'
        stanza_nlp_ru(my_text) 
        stanza_nlp_en(my_text)
```
