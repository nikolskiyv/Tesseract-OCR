```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

url = 'http://baza-r.ru/enterprises/549-80.html'
# задаем число больше номера первой страницы, для старта цикла
pages = 141
tmp1 = 80

f = open('name.txt', 'a')
for n in range(80, pages):
    print(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    items = soup.find_all('div', class_ = 'card_info')

    for i in items:
        itemName = i.find('strong').text
        print(itemName)
        f.write(itemName)
        f.write('\n')

    tmp = '-' + str(n)
    tmp1 += 1
    url = url.replace(tmp, '-' + str(tmp1))
f.close()    
```
