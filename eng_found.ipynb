```python
from docx import Document
import stanza
import zipfile
import xml.etree.ElementTree
import os
```

```python
dirname = "data/txt"
test = os.listdir(dirname)
```

```python
counter = 1

for item in test:
    filename = "data/txt/file" + str(counter) + ".txt"
    counter += 1
    print(filename)
    f = open(filename, "r")
    s = f.read()
    s = s.replace(",", " ")
    s = s.replace(":"," ")
    s = s.replace("."," ")
    s = s.replace("("," ")
    s = s.replace(")"," ")
    s = s.replace("/"," ")
    s = s.replace("«"," ")
    s = s.replace("»"," ")
    s = s.replace('"'," ")
    f.close()
    f = open(filename, "w")
    f.write(s)
    f.close() 
```

```python
counter = 1
eng = "" 

for item in test:
    filename = "data/txt/file" + str(counter) + ".txt"
    print(filename)
    f = open(filename, "r")
    counter += 1
    for line in f.readlines():
        for word in line.split(' '):
            try: 
                lang = word[0]
            except: 
                lang = 'Я'
                    
            if (lang >= 'a' and lang <= 'z') or (lang >= 'A' and lang <= 'Z'):
                if word not in eng:
                    eng += word + '\n'
    f.close()        
eng = eng.replace("\n", "")    
f = open("eng.txt", "w")
f.write(eng)
f.close
```
