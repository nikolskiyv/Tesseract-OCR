```python
import os

wordlistfile = open("wlf", "r")
wlf = wordlistfile.read()

rus = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪФЬЭЮЯабвгдеёжзийклмнопрстуфхцчшщъыьэюя" 
eng = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"

counter1 = 0
counter2 = 0

for file in os.listdir("before"):
    print(file)
    filename = "before/" + file
    
    if filename[-4:] == ".txt":
        before = open(filename, "r")
        string = before.read()
        for sym in string:
            if (sym not in rus) and (sym not in eng) and (sym != "\n"):
                string = string.replace(sym, " ")
        string = ' '.join(string.split())         

        for word in wlf.split():
            if word in string:
                counter1 += 1

        before.close()
        
for file in os.listdir("after"):
    print(file)
    filename = "after/" + file
    
    if filename[-4:] == ".txt":
        after = open(filename, "r")
        string = after.read()
        for sym in string:
            if (sym not in rus) and (sym not in eng) and (sym != "\n"):
                string = string.replace(sym, " ")
        string = ' '.join(string.split())         

        for word in wlf.split():
            if word in string:
                counter2 += 1

        before.close()        

print("Before: %d" % (counter1))        
print("After: %d" % (counter2))  
```
